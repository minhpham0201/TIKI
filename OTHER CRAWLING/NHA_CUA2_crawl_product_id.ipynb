```python
list_url = ['https://tiki.vn/den-cam-ung/c23242',
            'https://tiki.vn/den-diet-con-trung/c2020',
            'https://tiki.vn/den-xong-tinh-dau-tinh-dau-va-phu-kien/c23802',
            'https://tiki.vn/xich-du/c23424',
            'https://tiki.vn/ban-ghe/c23430',
            'https://tiki.vn/ke-trang-tri-cay/c23440',
            'https://tiki.vn/be-boi-va-phu-kien/c23434',
            'https://tiki.vn/dep-nha-tam/c27118',
            'https://tiki.vn/mieng-lot-bon-cau/c23260',
            'https://tiki.vn/rem-phong-tam/c23256',
            'https://tiki.vn/xo-thau-chau/c23834',
            'https://tiki.vn/bon-cau/c23270',
            'https://tiki.vn/lavabo/c23272',
            'https://tiki.vn/tui-giat/c23032',
            'https://tiki.vn/ban-de-ui-va-phu-kien/c23034',
            'https://tiki.vn/cay-lan-bui/c23036',
            'https://tiki.vn/bao-ho-lao-dong/c4452',
            'https://tiki.vn/xe-day-hang/c12984',
            'https://tiki.vn/da-cat/c23688',
            'https://tiki.vn/que-han/c23690',
            'https://tiki.vn/cua-luoi-cua/c23692',
            'https://tiki.vn/thang-nhom/c2327',
            'https://tiki.vn/bo-luc-giac/c4451',
            'https://tiki.vn/cay-lan-sung-phun-son/c23714',
            'https://tiki.vn/sung-ban-keo/c23712',
            'https://tiki.vn/may-do-khoang-cach/c27190',
            'https://tiki.vn/may-do-ph/c27192',
            'https://tiki.vn/may-do-nhiet-do-do-am/c27194',
            'https://tiki.vn/thiet-bi-do-luong-khac/c27196',
            'https://tiki.vn/thuoc-do-cac-loai/c2805',
            'https://tiki.vn/dan-piano/c23450',
            'https://tiki.vn/amply/c23462',
            'https://tiki.vn/dan-organ/c23464',
            'https://tiki.vn/dan-guitar-ukulele/c23466',
            'https://tiki.vn/phu-kien-dan/c23468',
            'https://tiki.vn/nhac-cu-dan-toc/c10076',
            'https://tiki.vn/ken-trong-phu-kien/c10074',
            'https://tiki.vn/phu-kien-khac/c23552',
            'https://tiki.vn/cac-loai-dan-khac/c10073'
    
            ]

list_page_50 = []
page_error = []

for url in list_url:
    parsed_url = urlparse(url)
    path_segments = parsed_url.path.split('/')

    category_name = path_segments[1]
    category_id = path_segments[2][1:]

    params = {
        'limit': '40',
        'include': 'advertisement',
        'aggregations': '2',
        'trackity_id': '74634137-8702-8b12-fe42-46a46a2a2573',
        'category': category_id,
        'page': '1',
        'urlKey': category_name
    }

    product_data = []
    
    response = requests.get('https://tiki.vn/api/personalish/v1/blocks/listings', headers=headers, params=params)
    
    if response.status_code == 200:
        response_json = response.json()
        last_page = response_json.get('paging', {}).get('last_page')
        print('total pages', last_page)
        print('CATEGORY', category_name)
        
        if last_page == 50:
            list_page_50.append(category_name)
            continue
        
        for i in range(1, last_page + 1):
            params['page'] = i
            response = requests.get('https://tiki.vn/api/personalish/v1/blocks/listings', headers=headers, params=params)

            if response.status_code == 200:
                print('request done page', i)
                response_json = response.json()
                for record in response_json.get('data', []):
                    product_id = record.get('id')
                    product_data.append({'product_id': product_id, 'category_name': category_name, 'max_page': last_page})
            else:
                print('cannot request page', i)
                page_error.append(i)
            time.sleep(random.randrange(3, 10))

        product_df = pd.DataFrame(product_data, columns=['product_id', 'category_name', 'max_page'])
        output_file_name = f'NHA-CUA2-{category_name}.csv'
        product_df.to_csv(output_file_name, index=False)
        
    else:
        print(f'Cannot request data for category {category_name}')
        page_error.append(category_name)

page_50_file = 'page50.txt'
cannot_request_page_file = 'page_error.txt'

# Open the file in append mode
with open(page_50_file, 'a') as file:
    # Write each item of the list to a separate line in the file
    file.write('\n'.join(list_page_50) + '\n')

with open(cannot_request_page_file, 'a') as file:
    # Write each item of the list to a separate line in the file
    file.write('\n'.join(page_error) + '\n')
```
